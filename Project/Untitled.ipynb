```python
df = pd.read_csv('weather_data_2sites-1.csv',parse_dates=['timestamp']).drop(columns =['temperature_site2','humidity_site2','Unnamed: 0'])
start_date ='2018-03-07 00:00:00'
end_date = '2018-03-10 00:00:00'
df= df[(df['timestamp'] >= start_date) & (df['timestamp'] < end_date)]
# df.set_index('timestamp',inplace=True,drop=True)
# df.plot(x='timestamp')
# df['date'] = df.index.date
# df['time'] = df.index.time
# df.set_index(['date','time'],inplace=True,drop=True)
# df=df.unstack('date')
df['date'],df['time'] = df["timestamp"].dt.day,df['timestamp'].dt.time
df.drop(columns = 'timestamp',inplace=True)
df.reset_index(inplace=True,drop=True)
# df=df.groupby('date')
k = pd.DataFrame()

for day in df['date'].unique():
    k[f'temperature_day{day}'] = df[df['date']==day]['temperature_site1'].values
    k[f'humidity_day{day}'] = df[df['date']==day]['humidity_site1'].values
pd.plotting.scatter_matrix(k,figsize=[20,20],diagonal='kde')
plt.show()
# k
```
